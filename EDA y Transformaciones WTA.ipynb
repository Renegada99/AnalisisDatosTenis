### Primeras Transformaciones - Manejo de Nulos

Los nulos que salgan en cada columna se convertiran en 0, esto con el fin de crear columnas calculadas necesarias para la identificación de patrones claves que influyan en el rendimiento de un deportista 

```python
# Reemplazar NaN por 0  
df_WTA['Best of'] = df_WTA['Best of'].fillna('0.0')
df_WTA['WRank'] = df_WTA['WRank'].fillna('0.0')
df_WTA['LRank'] = df_WTA['LRank'].fillna('0.0')
df_WTA['WPts'] = df_WTA['WPts'].fillna('0.0')
df_WTA['LPts'] = df_WTA['LPts'].fillna('0.0')
df_WTA['W1'] = df_WTA['W1'].fillna('0.0')
df_WTA['L1'] = df_WTA['L1'].fillna('0.0')
df_WTA['W2'] = df_WTA['W2'].fillna('0.0')
df_WTA['L2'] = df_WTA['L2'].fillna('0.0')
df_WTA['W3'] = df_WTA['W3'].fillna('0.0')
df_WTA['L3'] = df_WTA['L3'].fillna('0.0')

df_WTA['Wsets'] = df_WTA['Wsets'].fillna('0.0')
df_WTA['Lsets'] = df_WTA['Lsets'].fillna('0.0')
df_WTA['B365W'] = df_WTA['B365W'].fillna('0.0')
df_WTA['B365L'] = df_WTA['B365L'].fillna('0.0')

df_WTA['PSW'] = df_WTA['PSW'].fillna('0.0')
df_WTA['PSL'] = df_WTA['PSL'].fillna('0.0')
df_WTA['MaxW'] = df_WTA['MaxW'].fillna('0.0')
df_WTA['MaxL'] = df_WTA['MaxL'].fillna('0.0')
df_WTA['AvgW'] = df_WTA['AvgW'].fillna('0.0')
df_WTA['AvgL'] = df_WTA['AvgL'].fillna('0.0')
df_WTA['EXW'] = df_WTA['EXW'].fillna('0.0')
df_WTA['EXL'] = df_WTA['EXL'].fillna('0.0')
df_WTA['LBW'] = df_WTA['LBW'].fillna('0.0')
df_WTA['LBL'] = df_WTA['LBL'].fillna('0.0')
df_WTA['SJW'] = df_WTA['SJW'].fillna('0.0')
df_WTA['SJL'] = df_WTA['SJL'].fillna('0.0')
```

### Creación de las columnas calculadas

Las columnas calculadas están dadas por los siguientes items de consideración:

- Puntuación total del partido
- Rendimiento Ajustado por Ranking
- Efectividad en Sets y Juegos Ganados
- Impacto de la Probabilidad de Apuesta

```python
# Ecuación 1: P_Total
df_WTA['P_Total'] = (
    df_WTA['Wsets'] * 10 +
    df_WTA[['W1', 'W2', 'W3']].sum(axis=1) -
    df_WTA[['L1', 'L2', 'L3']].sum(axis=1)
)
df_WTA
```

```python
# Ecuación 2: Ranking Ajustado
df_WTA['R_Ajustado'] = df_WTA['WRank'] / (df_WTA['LRank'] + 1)
df_WTA
```

```python
# Ecuación 3: Probabilidad de impacto de apuesta (P_Apuesta)
df_WTA['P_Apuesta'] = 1 / (df_WTA['B365W'] + df_WTA['EXW'] + df_WTA['LBW'] + df_WTA['PSW'] + df_WTA['SJW']) 
df_WTA
```

```python
print(df_WTA[['B365W', 'EXW', 'LBW', 'PSW', 'SJW']].dtypes)
```

```python
# Ecuación 4: Efectividad en Juegos
cols_wsets = ['W1', 'W2', 'W3']
cols_lsets = ['L1', 'L2', 'L3']
df_WTA['E_Juegos'] = df_WTA[cols_wsets].sum(axis=1) / (df_WTA[cols_lsets].sum(axis=1) + 1)
df_WTA
```

Se crea la variable objetivo denominada Ganador:

```python
df_WTA['Ganador'] = (df_WTA['Wsets'] > df_WTA['Lsets']).astype(int)
df_WTA
```
